```python
df.loc[
    (df["Aptitude_Result"] == "Fail"),
    "Technical_Score"
] = np.nan
```

```python
df.loc[
    (df["Technical_Result"] == "Fail"),
    "HR_Score"
] = np.nan
```

```python
df.loc[
    df["Offer_Released"] == "No",
    "Offer_Accepted"
] = "No"
```

```python
df["Time_Gap_Days"].fillna(
    df["Time_Gap_Days"].median(),
    inplace=True
)
```

```python
df.isnull().sum()
```

```python
total_candidates = df.shape[0]

aptitude_pass = df[df["Aptitude_Result"] == "Pass"].shape[0]
technical_pass = df[df["Technical_Result"] == "Pass"].shape[0]
hr_pass = df[df["HR_Result"] == "Pass"].shape[0]
offers_released = df[df["Offer_Released"] == "Yes"].shape[0]
offers_accepted = df[df["Offer_Accepted"] == "Yes"].shape[0]

total_candidates, aptitude_pass, technical_pass, hr_pass, offers_released, offers_accepted
```

```python
funnel = pd.DataFrame({
    "Stage": [
        "Applied",
        "Aptitude Passed",
        "Technical Passed",
        "HR Passed",
        "Offer Released",
        "Offer Accepted"
    ],
    "Count": [
        total_candidates,
        aptitude_pass,
        technical_pass,
        hr_pass,
        offers_released,
        offers_accepted
    ]
})

funnel["Conversion_%"] = funnel["Count"] / total_candidates * 100
funnel
```
